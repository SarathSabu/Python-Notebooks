* It seems like a question (QID) may receive responses from multiple authors. Can you create a dictionary that keeps the number of responses each question received?

````
# Can you create a dictionary that keeps the number of responses each question received?
QA_Num_of_Responses = {}
for qid in set(QIDs):
  QA_Num_of_Responses[qid] = len(QA[QA['QID'] == qid])
print(QA_Num_of_Responses)
````

````
import collections
from collections import Counter
Counter = Counter(QIDs)
print(Counter)
````

````
# Which question received the most responses, and how many responses did it receive?
[qid for qid in set(QIDs) if QA_Num_of_Responses[qid] == max(QA_Num_of_Responses.values())]
````

````
# Is there another way to identify this question ID?
max_QID = max(QA_Num_of_Responses, key=QA_Num_of_Responses.get)
print(max_QID)
````

* ```key=QA_Num_of_Responses.get```  is the crucial part in the command. It specifies a custom function ```QA_Num_of_Responses.get``` to be used by ```max()``` to determine the maximum. The ```.get``` method of a dictionary returns the value associated with a given key. In this case, for each ```QID (key)``` in the dictionary, ```QA_Num_of_Responses.get``` will return the number of responses (value).

* However, the ```max()``` function only returns the **first** key value it encounters that is associated with the maximum value.

* This limitation is manifested  clearly in the following example where we seek to find questions that received the least amount of responses.



````
# The least amount of response to keep a question in our dataset
print(min(QA_Num_of_Responses.values()))
````

````
# Which question received the most responses, and how many responses did it receive?
[qid for qid in set(QIDs) if QA_Num_of_Responses[qid] == min(QA_Num_of_Responses.values())]
````

* It seems that many questions received the least amount of responses, which is one in our dataset.

* Only one question ID is returned by the built in ```min()``` function, though the execution is much faster.

````
# Is there another way to find out this question?
min_QID = min(QA_Num_of_Responses, key=QA_Num_of_Responses.get)
print(min_QID)
````

#### Loop through Python DataFrame

Loops are fundamental methods that repeat a block of code multiple times. There are mainly two types of loops in Python, the ```for``` lop and the ``` while ``` loop.

1. ```for``` loop: Used to iterate over a sequence (like a list, tuple, string, or range) or other iterable objects.

2. ```while``` loop: Used to repeat a block of code as long as a condition is True.


````
# Let's identify all questions that contain the keyword "boss", which might indicate
# that the question is related to career development.

# To avoid repeatitive processing of the same question, we first identify all the unique questions in the dataset.
unique_questions = QA[['QID', 'Question_English']].drop_duplicates()
career_questions = []
for i in range(unique_questions.shape[0]):
    if 'boss' in unique_questions['Question_English'].iloc[i].lower():
        career_questions.append(unique_questions[['QID', 'Question_English']].iloc[i])
print(len(career_questions))
````

* As we introduced earlier, ```.lower()``` is a method that operates on strings (text data). Its purpose is to convert any uppercase letters within the string to their lowercase equivalents.

````
# We can identify the same set of questions using a while loop.
career_questions = []
qcount=0
while qcount < unique_questions.shape[0]:
    if 'boss' in unique_questions['Question_English'].iloc[qcount].lower():
        career_questions.append(unique_questions[['QID', 'Question_English']].iloc[qcount])
    qcount += 1
print(len(career_questions))
````

* ```qcount = 0``` A counter variable qcount is initialized to 0. This variable will keep track of the current question being examined.

````
# Check if any missing values exist in each column
missing_values_per_column = QA.isna().any(axis = 0)
print(missing_values_per_column)
````

* ```QA.isna()``` This part applies the ```.isna()``` method to the DataFrame QA. The ```.isna()``` method checks each individual cell within the DataFrame. If a cell is empty or contains a missing value (often represented as NaN which stands for "Not a Number"), it's marked as True; otherwise, it's marked as False. The result is a new DataFrame with the same shape as QA, but with True/False values indicating missing data.

* The ```.any()``` method is then chained to the result of ```QA.isna()```. The ```.any()``` method is applied column-wise (because of the default axis=0 behavior) to the True/False DataFrame. For each column, if there is at least one True value (meaning at least one missing value in that column), ```.any()``` will return True for that column. Otherwise, it returns False.

````
# Check if any missing values exist in each row
missing_values_per_row = QA.isna().any(axis=1)
print(missing_values_per_row)
````

````
# Check for missing values and summarize
missing_values = QA.isna()
missing_values_per_column = missing_values.any()
total_missing_values = missing_values.sum().sum()

print("Missing values per column:\n", missing_values_per_column)
print("\nTotal missing values:", total_missing_values)
````

* Scenario 2: Create a new column 'QuestionType' that indicates the type of question based on keywords in the 'Question_English' column. For instance, 'Career' if the question contains "job" or "career", 'Relationship' if it contains "relationship" or "family", and 'MentalHealth' otherwise.

````
QA['QuestionType'] = QA['Question_English'].apply(lambda x: 'Career' if any(keyword in x.lower() for keyword in ['job', 'career']) else 'Relationship' if any(keyword in x.lower() for keyword in ['relationship', 'family']) else 'MentalHealth')
print(QA[['QuestionType']])
````

* Scenario 3: Create a new column 'BonusAndUsefulness' that combines the information from 'Received_Bonus_Yes_No' and 'Usefulness_vote'. For example, 'Bonus_HighUsefulness' if the answer received a bonus and has a high usefulness vote, 'NoBonus_HighUsefulness' if it didn't receive a bonus but has a high usefulness vote, and so on.

````
QA['BonusAndUsefulness'] = QA.apply(lambda row: f"{row['Received_Bonus_Yes_No']}_{'HighUsefulness' if row['Usefulness_vote'] > 5 else 'LowUsefulness'}", axis=1)
print(QA['BonusAndUsefulness'])
````
